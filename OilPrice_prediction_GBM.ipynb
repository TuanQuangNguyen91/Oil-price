## Brent Oil's price prediction using Geometric Brownian Motion

```python
# Import libraries
import matplotlib.pyplot as plt
import warnings
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
plt.style.use('seaborn')
# plt.style.use('seaborn-colorblind') #alternative
plt.rcParams['figure.figsize'] = [15, 7]
plt.rcParams['figure.dpi'] = 300
warnings.simplefilter(action='ignore', category=FutureWarning)

# Import data
df = pd.read_csv(r'Europe_Brent_Spot_Price_FOB.csv', index_col = 'Day', parse_dates = True)
df.index = pd.to_datetime(df.index)
df.columns = ['Brent_price']
df['Price'] = 1
df.sort_index(inplace = True)
df.head()
```

```python
data = df['Brent_price']
returns = data.pct_change().dropna()

ax = returns.plot()
ax.set_title('Brent Oil price returns', fontsize=16)
plt.show()
```

```python
#Create train and test sets
train = returns['2018':'2021']
test = returns['2022-01-01':'2022-02-01']

#Parameters for simulations
T = len(test)
N = len(test)
S_0 = data[train.index[-1]]
N_SIM = 100
mu = train.mean()
sigma = train.std()

# Define function for simulation
def simulate_gbm(s_0, mu, sigma, n_sims, T, N, random_seed=42):
    np.random.seed(random_seed)
    dt = T/N  #time increment
    dW = np.random.normal(scale = np.sqrt(dt), size=(n_sims, N)) #Brownian increment
    W = np.cumsum(dW, axis=1)  #Brownian motion
    time_step = np.linspace(dt, T, N)  
    time_steps = np.broadcast_to(time_step, (n_sims, N))
    S_t = s_0 * np.exp((mu - 0.5 * sigma ** 2) * time_steps + sigma * W)
    S_t = np.insert(S_t, 0, s_0, axis=1)
    
    return S_t

# Run simulation
gbm_simulations = simulate_gbm(S_0, mu, sigma, N_SIM, T, N)

# Plot the simulation results
last_train_date = train.index[-1].date()
first_test_date = test.index[0].date()
last_test_date = test.index[-1].date()
plot_title = (f'Brent Oil price Simulation ({first_test_date}:{last_test_date})')

selected_indices = data[last_train_date:last_test_date].index
index = [date.date() for date in selected_indices]

#Using transpose to have dimension (timestep x simulations)
gbm_simulations_df = pd.DataFrame(np.transpose(gbm_simulations), index=index)

# Plotting
fig, ax = plt.subplots(figsize = (15,10))
gbm_simulations_df.plot(alpha=0.2, legend=False, ax =ax)
line_1, = ax.plot(index, gbm_simulations_df.mean(axis=1), color='red')
line_2, = ax.plot(index, data[last_train_date:last_test_date], color='blue')
ax.set_title(plot_title, fontsize=16)
ax.legend((line_1, line_2), ('mean', 'actual'))
plt.show()
```

```python
train_2 = returns['2022-01':'2022-06']
test_2 = returns['2022-7']

T2 = len(test_2)
N2 = len(test_2)
S_02 = data[train_2.index[-1]]
N_SIM = 100
mu2 = train_2.mean()
sigma2 = train_2.std()

# Run simulation
gbm_simulations_2 = simulate_gbm(S_02, mu2, sigma2, N_SIM, T2, N2)

# Plot the simulation results
last_train_date2 = train_2.index[-1].date()
first_test_date2 = test_2.index[0].date()
last_test_date2 = test_2.index[-1].date()
plot_title2 = (f'Brent Oil price Simulation ({first_test_date2}:{last_test_date2})')

selected_indices2 = data[last_train_date2:last_test_date2].index
index2 = [date.date() for date in selected_indices2]

#Using transpose to have dimension (timestep x simulations)
gbm_simulations_df2 = pd.DataFrame(np.transpose(gbm_simulations_2), index=index2)

# Plotting
fig, ax = plt.subplots(figsize = (15,10))
gbm_simulations_df2.plot(alpha=0.2, legend=False, ax = ax)
line_1, = ax.plot(index2, gbm_simulations_df2.mean(axis=1), color='red')
line_2, = ax.plot(index2, data[last_train_date2:last_test_date2], color='blue')
ax.set_title(plot_title2, fontsize=16)
ax.legend((line_1, line_2), ('mean', 'actual'))
plt.show()
```
